```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
import pandas as pd

filename="./GLMCC/results/J_62_0.5Hz-w.csv"
df = pd.read_csv(filename)
data = df.values
matrix = np.zeros((106, 106))

for i in range(data.shape[0]):
    R = int(data[i, 0])
    T = int(data[i, 1])
    if np.abs(data[i, 2]) > data[i, 4]:
        matrix[R, T] = 10

    if np.abs(data[i, 3]) > data[i, 5]:
        matrix[T, R] = 10

ca1_dg = matrix[0: 55, 73: 105]
ca1_ca3 = matrix[0: 55, 56: 72]
ca3_dg = matrix[56: 72, 73: 105]
ca3_ca1 = matrix[56: 72, 0: 55]
dg_ca1 = matrix[73: 105, 0: 55]
dg_ca3 = matrix[73: 105, 56: 72]

fig = plt.subplot(2,3,1)
fig.matshow(ca1_dg)
fig.set_xticks([])
fig.set_yticks([])
fig.set_title('CA1 -> DG', fontsize = 9, y = .96)
fig.set_xlabel('DG', fontsize=8)
fig.set_ylabel('CA1', fontsize=8)

fig = plt.subplot(2,3,2)
fig.matshow(ca1_ca3)
fig.set_xticks([])
fig.set_yticks([])
fig.set_title('CA1 -> CA3', fontsize = 9, y = .96)
fig.set_xlabel('CA3', fontsize=8)
fig.set_ylabel('CA1', fontsize=8)

fig = plt.subplot(2,3,3)
fig.matshow(ca3_dg)
fig.set_xticks([])
fig.set_yticks([])
fig.set_title('CA3 -> DG', fontsize = 9, y = .96)
fig.set_xlabel('DG', fontsize=8)
fig.set_ylabel('CA3', fontsize=8)

fig = plt.subplot(2,3,5)
fig.matshow(ca3_ca1.T)
fig.set_xticks([])
fig.set_yticks([])
fig.set_title('CA3 -> CA1', fontsize = 9, y = .96)
fig.set_xlabel('CA3', fontsize=8)
fig.set_ylabel('CA1', fontsize=8)

fig = plt.subplot(2,3,4)
fig.matshow(dg_ca1.T)
fig.set_xticks([])
fig.set_yticks([])
fig.set_title('DG -> CA1', fontsize = 9, y = .96)
fig.set_xlabel('DG', fontsize=8)
fig.set_ylabel('CA1', fontsize=8)


fig = plt.subplot(2,3,6)
fig.matshow(dg_ca3.T)
fig.set_xticks([])
fig.set_yticks([])
fig.set_title('DG -> CA3', fontsize = 9, y = .96)
fig.set_xlabel('DG', fontsize=8)
fig.set_ylabel('CA3', fontsize=8)

plt.savefig('a.png',dpi=700)
plt.margins(0, 0)
plt.show()
```
